```python
# Create a new column in the DataFrame (1) - Price difference
MS['PriceDiff'] = MS['Close'].shift(-1) - MS['Close']
```

```python
MS.head()
```

```python
# SHIFT ALL PRICE OF CLOSE COLUMN ONE ROW UPWARD AND MAKE A NEW COLUMN #
MS['price'] = MS['Close'].shift(-1)
```

```python
MS.head()
```

```python
MS.tail()
```

```python
# Create a new column in the DataFrame (2) - Daily return

# Daily Return is calcuated as PriceDiff/Close
#Create a new column Return in the DataFrame MS
MS['Daily Return'] = MS['PriceDiff'] /MS['Close']
```

```python
MS.head()
```

```python
print(MS['Daily Return'].loc['2017-08-21'])
```

```python
MS['Direction'] = [1 if MS['PriceDiff'].loc[var] > 0 else -1 for var in MS.index]
```

```python
MS.head()
```

```python
###MS[' Moveing Average'] =(MS['Close']+MS['Close'].shift(1)+MS['Close'].shift(2))/3###
#logic behind rolling func of python

MS['MA40'] = MS['Close'].rolling(40).mean()
MS=MS.dropna()
#plot the moving average
plt.figure(figsize=(16, 8))
MS['MA40'].loc['2017-01-01':'2017-12-31'].plot(label='MA40')
MS['Close'].loc['2017-01-01':'2017-12-31'].plot(label='Close')
plt.legend()
plt.show()
```

```python
MS.tail()
```

```python
plt.figure(figsize=(10, 8))
MS['MA300'] = MS['Close'].rolling(300).mean()
MS=MS.dropna()
MS['MA40'].plot()
MS['MA300'].plot()
MS['Close'].plot()
plt.legend()
```

```python
MS.tail()
```

```python
MS['SHARES']= [1 if MS.loc[var,'MA40']>MS.loc[var,'MA300'] else 0 for var in MS.index]
MS=MS.dropna()
```
